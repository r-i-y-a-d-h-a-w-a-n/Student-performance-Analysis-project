```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
df=pd.read_csv('StudentsPerformance.csv')
print(df)
```

```python
plt.figure(figsize=(6,6))
plt.title("Gender distribution")
a=sns.countplot(x="gender", data=df, palette='Accent')
a.bar_label(a.containers[0]) #this line give the exact count of male and female distribution
plt.show()
```

```python
b= df.groupby("parental level of education").agg({"math score": "mean", "reading score": "mean", "writing score":"mean"})
sns.heatmap(b,annot=True)
plt.show() # education of parent have good impact in childrens
```

```python
c=sns.countplot(x="test preparation course", data=df)
d=c.bar_label(c.containers[0])
plt.show() #According  to test preparation course none prepared course is greater than completed course
```

```python
plt.figure(figsize=(5,5))
b=df.groupby("reading score").agg({"math score":'mean', "reading score":"mean"})
sns.heatmap(b)
plt.show()
```

```python
sns.boxplot(x="reading score", data=df)
#sns.boxplot(x="math score", data=df)
plt.show()
```

```python
b=sns.countplot(x="race/ethnicity", data=df)
b.bar_label(b.containers[0])
plt.show() #Ethnic group D has more distribution
```

```python
a=pd.Series(df['race/ethnicity']).unique()
g1=df.loc[df['race/ethnicity']=='group A'].count()
g2=df.loc[df['race/ethnicity']=='group B'].count()
g3=df.loc[df['race/ethnicity']=='group C'].count()
g4=df.loc[df['race/ethnicity']=='group D'].count()
g5=df.loc[df['race/ethnicity']=='group E'].count()
res= [g1['gender'], g2['gender'], g3['gender'],g4['gender'],g5['gender']]
plt.pie(res, labels=a,  autopct="%1.2f%%")
plt.show()
```
